We iterate through the indices of the 'BMI' values in the 'data' dictionary using a for loop.

We use square brackets to access the 'BMI' values at each index.

We compare the 'BMI' value at the current index to categorize it based on your criteria which 
bmi less than 185 is underweight
 18.5 between 24.9 is normal
25.0 to 29.9 overweight
30 and greater is obessed

We update the 'BMI' value in the 'data' dictionary at the current index to the corresponding category.

This code will categorize the 'BMI' values in your 'data' dictionary and update them in place.

```python
for i in range(len(data['BMI'])):
    if data['BMI'][i] < 18:
        data['BMI'][i] = 'Underweight'
    elif  data['BMI'][i]>=18 and  data['BMI'][i]< 25:
        data['BMI'][i] = 'Normal'
    elif data['BMI'][i] ==25 or data['BMI'][i]<= 30:
        data['BMI'][i] = 'overweight'
    else:
       data['BMI'][i]  = 'obbessed'
```
